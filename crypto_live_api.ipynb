```python
import requests
import pandas as pd
from datetime import datetime

def fetch_crypto_data():
    print("📡 Pinging CoinGecko REST API...")

    url = "https://api.coingecko.com/api/v3/coins/markets"

    params = {
        "vs_currency": "usd",
        "order": "market_cap_desc",
        "per_page": 50,
        "page": 1,
        "sparkline": False
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        raw_json_data = response.json()
        print(f"✅ Successfully retrieved {len(raw_json_data)} assets.")

        df = pd.DataFrame(raw_json_data)

        # Added .copy() here to eliminate the Pandas warning
        clean_df = df[['symbol', 'name', 'current_price', 'market_cap', 'total_volume', 'price_change_percentage_24h']].copy()

        clean_df['last_updated'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        clean_df.to_csv("live_crypto_market_data.csv", index=False)
        print("💾 Data cleaned and saved to 'live_crypto_market_data.csv'")

        return clean_df
    else:
        print(f"❌ API Request Failed. Status Code: {response.status_code}")

live_data = fetch_crypto_data()
display(live_data.head())
```
