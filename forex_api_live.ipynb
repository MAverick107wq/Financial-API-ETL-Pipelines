```python
import requests
import pandas as pd

def fetch_forex_rates():
    print("📡 Pinging Frankfurter Forex API...")

    # =====================================================================
    # 1. API CONFIGURATION
    # =====================================================================
    url = "https://api.frankfurter.app/latest"
    params = {"from": "USD"}

    # =====================================================================
    # 2. EXECUTE GET REQUEST
    # =====================================================================
    response = requests.get(url, params=params)

    if response.status_code == 200:
        raw_json_data = response.json()

        # =====================================================================
        # 3. NESTED JSON PARSING
        # =====================================================================
        rates_dict = raw_json_data.get("rates", {})
        print(f"✅ Successfully retrieved exchange rates for {len(rates_dict)} fiat currencies.")

        # =====================================================================
        # 4. DATAFRAME TRANSFORMATION
        # =====================================================================
        df = pd.DataFrame(list(rates_dict.items()), columns=["Target_Currency", "Exchange_Rate"])

        # Append metadata from the parent JSON level
        df["Base_Currency"] = raw_json_data.get("base")
        df["Last_Updated"] = raw_json_data.get("date")

        # Reorder columns for optimal readability
        clean_df = df[['Base_Currency', 'Target_Currency', 'Exchange_Rate', 'Last_Updated']].copy()

        # =====================================================================
        # 5. DATA EXPORT
        # =====================================================================
        clean_df.to_csv("live_forex_usd_rates.csv", index=False)
        print("💾 Data cleaned and saved to 'live_forex_usd_rates.csv'")

        return clean_df
    else:
        print(f"❌ API Request Failed. Status Code: {response.status_code}")

# Execute Pipeline
forex_data = fetch_forex_rates()
display(forex_data.head())
```
